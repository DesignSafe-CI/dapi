# Projects on DesignSafe

This notebook demonstrates how to list, inspect, and access files in DesignSafe projects using `dapi`.

```python
from dapi import DSClient

ds = DSClient()
```

## List Your Projects

```python
projects = ds.projects.list()
print(f"You have access to {len(projects)} projects\n")
for p in projects[:10]:
    pi_name = f"{p['pi']['fname']} {p['pi']['lname']}" if p["pi"] else "N/A"
    print(f"{p['projectId']:12s} | {pi_name:25s} | {p['title'][:50]}")
```

## Get Project Details

```python
info = ds.projects.get("PRJ-6270")
print(f"Title:       {info['title']}")
print(f"Type:        {info['projectType']}")
print(f"Description: {info['description'][:100]}..." if info["description"] else "")
print(f"PI:          {info['pi']['fname']} {info['pi']['lname']}" if info["pi"] else "")
print(f"DOIs:        {info['dois']}")
print(f"Keywords:    {info['keywords']}")
print(f"System ID:   {info['systemId']}")
```

## List Files in a Project

```python
files = ds.projects.files("PRJ-6270")
for f in files:
    print(f"{f.name:50s} {f.type}")
```

```python
# List files in a subfolder
files = ds.projects.files("PRJ-1305", path="/Training/")
for f in files[:10]:
    print(f"{f.name:50s} {f.type}")
```

## Use Project Paths with File Operations

`ds.files.to_uri()` also accepts project paths and resolves the PRJ number automatically.

```python
# Translate a project path to a Tapis URI
uri = ds.files.to_uri("/MyProjects/PRJ-6270/data/")
print(f"Tapis URI: {uri}")

# List files using the URI
files = ds.files.list(uri)
for f in files:
    print(f"  {f.name} ({f.type})")
```
