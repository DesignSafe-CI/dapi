# Projects on DesignSafe

This notebook demonstrates how to list, inspect, and access files in DesignSafe projects using `dapi`.

```python
from dapi import DSClient

ds = DSClient()
```

## List Your Projects

Returns a DataFrame by default. Use `output="list"` for a list of dicts.

```python
ds.projects.list()
```

## Get Project Details

```python
info = ds.projects.get("PRJ-6270")
info
```

## List Files in a Project

Returns a DataFrame by default. Use `output="raw"` for Tapis file objects.

```python
ds.projects.files("PRJ-6270")
```

```python
# List files in a subfolder
ds.projects.files("PRJ-1305", path="/Training/")
```

## Use Project Paths with File Operations

`ds.files.to_uri()` also accepts project paths and resolves the PRJ number automatically.

```python
# Translate a project path to a Tapis URI
uri = ds.files.to_uri("/MyProjects/PRJ-6270/data/")
print(f"Tapis URI: {uri}")

# List files using the URI
files = ds.files.list(uri)
for f in files:
    print(f"  {f.name} ({f.type})")
```
