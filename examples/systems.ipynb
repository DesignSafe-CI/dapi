# Systems on DesignSafe

This notebook shows how to list available HPC and storage systems, check TMS credentials, and view queue information.

```python
from dapi import DSClient

ds = DSClient()
```

## List All Systems (HPC + Storage)

```python
ds.systems.list()
```

## HPC Execution Systems Only

Shows credential status: `True` means TMS credentials are active.

```python
ds.systems.list("hpc")
```

## Storage Systems Only

```python
ds.systems.list("storage")
```

## Queue Information

View available queues for a specific HPC system.

```python
ds.systems.queues("stampede3")
```

## Manage TMS Credentials

```python
# Check credentials on a specific system
ds.systems.check_credentials("stampede3")
```
