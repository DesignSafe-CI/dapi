# File Access on DesignSafe

This notebook demonstrates how to access files across DesignSafe storage systems using `dapi`.

## Storage Areas

| JupyterHub Folder | Description |
|---|---|
| **MyData** | Personal files |
| **MyProjects** | Shared project files (team collaboration) |
| **CommunityData** | Public shared datasets (read-only) |
| **NHERI-Published** | Archived NHERI datasets with DOIs (read-only) |
| **NEES** | Legacy NEES datasets (read-only) |
| **Work** | Active HPC job I/O |

```python
from dapi import DSClient

ds = DSClient()
```

## Path Translation

`dapi` translates DesignSafe paths (as seen in Jupyter) to Tapis URIs automatically.

```python
# MyData
uri = ds.files.to_uri("/MyData/folder/")
print("MyData:", uri)
```

```python
# CommunityData
uri = ds.files.to_uri("/CommunityData/app_examples/")
print("CommunityData:", uri)
```

```python
# NHERI-Published
uri = ds.files.to_uri("/NHERI-Published/PRJ-1271/")
print("NHERI-Published:", uri)
```

```python
# NEES
uri = ds.files.to_uri("/NEES/")
print("NEES:", uri)
```

```python
# Projects - dapi resolves the PRJ number to the Tapis system UUID
uri = ds.files.to_uri("/MyProjects/PRJ-1305/")
print("MyProjects:", uri)
```

## Listing Files

```python
# List files in CommunityData
uri = ds.files.to_uri("/CommunityData/")
files = ds.files.list(uri)
for f in files[:10]:
    print(f"{f.name:50s} {f.type}")
```

```python
# List files in a project
uri = ds.files.to_uri("/MyProjects/PRJ-1305/Training/")
files = ds.files.list(uri)
for f in files[:10]:
    print(f"{f.name:50s} {f.type}")
```

## Reverse Translation (Tapis URI to local path)

```python
# Convert Tapis URIs back to JupyterHub paths
print(ds.files.to_path("tapis://designsafe.storage.default/kks32/results/"))
print(ds.files.to_path("tapis://designsafe.storage.community/app_examples/"))
print(ds.files.to_path("tapis://designsafe.storage.published/PRJ-1271/"))
print(ds.files.to_path("tapis://nees.public/NEES-2011-1050.groups/"))
```
