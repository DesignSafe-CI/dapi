# Published Datasets on DesignSafe

This notebook demonstrates how to search, inspect, and access published datasets on DesignSafe using `dapi`.

```python
from dapi import DSClient

ds = DSClient()
```

## List Recent Publications

Returns a DataFrame by default.

```python
ds.publications.list(limit=10)
```

## Search Publications

```python
# Search by topic
ds.publications.search("liquefaction")
```

```python
# Search by PI name
ds.publications.search("Rathje")
```

## Get Publication Details

```python
info = ds.publications.get("PRJ-6270")
info
```

## List Files in a Published Dataset

```python
ds.publications.files("PRJ-1271")
```
